```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
data = pd.read_csv(r'C:\Users\Tunzale\Desktop\DSDS\yeni\cluster (1).csv')
data
```

```python
data.drop(['CUST_ID'],axis=1,inplace=True)
```

```python
data.isnull().sum()
```

```python
data['MINIMUM_PAYMENTS'].fillna(data['MINIMUM_PAYMENTS'].mean(), inplace=True)
data['CREDIT_LIMIT'].fillna(data['CREDIT_LIMIT'].mean(), inplace=True)

```

```python
data.isnull().sum()
```

```python
from sklearn.preprocessing import normalize
data_scaled = normalize(data)
data_scaled = pd.DataFrame(data_scaled, columns=data.columns)
data_scaled.head()
```

```python
x=data[['BALANCE','BALANCE_FREQUENCY']]
x
```

```python
from sklearn.cluster import KMeans
wcss=[]
for i in range(1,7):
    kmeans = KMeans(i)
    kmeans.fit(x)
    wcss_iter = kmeans.inertia_
    wcss.append(wcss_iter)
```

```python
wcss
```

```python
number_clusters = range(1,7)
plt.plot(number_clusters,wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('Within-cluster Sum of Squares')
```

```python
kmeans = KMeans(5)
kmeans.fit(x)
```

```python
identified_clusters = kmeans.fit_predict(x)
identified_clusters
```

```python
data_with_clusters = data.copy()
data_with_clusters['Cluster'] = identified_clusters
data_with_clusters
```

```python
plt.scatter(data_with_clusters['BALANCE'],
            data_with_clusters['BALANCE_FREQUENCY'],c=data_with_clusters['Cluster'],cmap='rainbow')
plt.xlim()
plt.ylim()
plt.show()
```

```python
data['dependent_value'] = y
data
```

```python
X = data_scaled.copy()
y = data['dependent_value']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import roc_auc_score

base_model = RandomForestClassifier(n_estimators=100, random_state=42)
base_model.fit(X_train, y_train)

y_pred = base_model.predict(X_test)
y_prob = base_model.predict_proba(X_test)

roc_pred = roc_auc_score(y_test, y_prob, multi_class='ovr')
roc_prob = roc_auc_score(y_test, y_prob, multi_class='ovr')
gini_pred = roc_pred * 2 - 1

print("ROC AUC (Predicted):", roc_pred)
print("ROC AUC (Probabilities):", roc_prob)

```

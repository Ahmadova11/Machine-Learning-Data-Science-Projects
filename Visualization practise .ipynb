```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.linspace(0,5,10)
y=x*x
x
y
```

```python
plt.plot(x, y,'r')
plt.xlabel('my love')
plt.ylabel('days')
plt.title('my love for Nizi')
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r--')
plt.subplot(1,2,1)
plt.plot(y,x,'g*-')
```

```python
x=np.arange(0,10,0.1)
y=np.sin(x)
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r--')
plt.subplot(1,2,2)
plt.plot(y,x,'b*-')
```

```python
fig=plt.figure()
axes=fig.add_axes([0.1,0.1,1,0.5])
axes.plot(x,y,'b')
axes.set_xlabel('X label')
axes.set_ylabel('Y label')
axes.set_title('Mustevi fiquru')
```

```python
fig=plt.figure()
axes1=fig.add_axes([0.1, 0.1, 1.8, 0.8])
axes2=fig.add_axes([0.8,0.5,0.4,0.3])

axes1.plot(x,y,'b')
axes1.set_xlabel('X_label')
axes1.set_ylabel('Y_label')
axes1.set_title('Boyukfigure')

axes2.plot(y,x,'r')
axes2.set_xlabel('X_label')
axes2.set_ylabel('Y_label')
axes2.set_title('Balacafigure');
```
